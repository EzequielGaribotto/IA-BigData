# Ejercicio 6: Creación y filtrado de DataFrames en R
En este notebook se realiza la creación de un DataFrame básico en R sobre productos y se filtran sus registros por categoría.

### Crea un DataFrame en R con columnas Nombre, Categoría y Precio.

```r
# Creamos un dataframe directamente con las funciones de R nativas
datos_productos <- data.frame(
  Nombre = c(
    "Portátil", "Televisor OLED", "Sofá 3 plazas",
    "Auriculares Inalámbricos", "Mesa de comedor", "Smartwatch"
  ),
  Categoria = c(
    "Electrónica", "Electrónica", "Mobiliario",
    "Electrónica", "Mobiliario", "Electrónica"
  ),
  Precio = c(1250.00, 950.50, 450.00, 89.99, 150.00, 210.00)
)
# Lo mostramos por pantalla
print(datos_productos)
```

### Filtra los productos que pertenecen a una categoría específica.

```r
# Inicializamos la variable categoria_deseada
categoria_deseada <- "Electrónica"

# Inicializamos productos_filtrados con un subset de los datos originales,
# los cuales vamos a filtrar por su columna "Categoría"
productos_filtrados <- subset(datos_productos, Categoria == categoria_deseada)

print(productos_filtrados)
```

### Calcula y muestra el precio promedio de los productos de esa categoría.


```r
# Utilizamos la función mean para calcular la media aritmética 
# de la columna Precio de los productos filtrados
precio_promedio <- mean(productos_filtrados$Precio)

# Mostramos el precio promedio por pantalla
cat(sprintf("El precio promedio de los productos en la categoría '%s' es: %.2f\n",
            categoria_deseada, precio_promedio))
```
