### Ejercicio 5: Crea un DataFrame sencillo utilizando la librería Pandas en Python.
Instrucciones:
1. Crea un DataFrame con datos de estudiantes: Nombre, Edad, y Nota.

```python
import pandas as pd # Importamos pandas, esencial para crear el dataframe
# Inicializamos la variable data en un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Marta', 'Pedro', 'Sofia'],
    'Edad': [20, 21, 19, 22, 20, 21],
    'Nota': [8.5, 4.0, 6.5, 9.0, 3.5, 5.0]
}
# Usamos pandas para crear el dataframe de nuestro diccionario
df = pd.DataFrame(data)

# Mostramos el dataframe por pantalla
print("--- DataFrame Inicial ---")
print(df)
```

### 2. Agrega una nueva columna Aprobado que indique si la nota es mayor o igual a 5.

```python
# Utilizamos una operación de filtro sobre la columna "Nota"
# para crear la columna "Aprobado", en el dataframe
#  con notas superiores o iguales a 5
df['Aprobado'] = df['Nota'] >= 5

# Mostramos el dataframe filtrado por pantalla
print("--- DataFrame con columna 'Aprobado' ---")
print(df)
```

### 3. Filtra los estudiantes que han aprobado.

```python
# Del dataframe original, filtramos aquellos
# que tengan la columna "Aprobado" en True
df_est_aprobados = df[df['Aprobado'] == True] # "== True" es redundante

# Mostramos el df filtrado por pantalla
print("--- Estudiantes Aprobados ---")
print(df_est_aprobados)
```

### 4. Calcula y muestra el promedio de las notas.

```python
# Utilizamos la función "mean" de pandas para calcular
# la media aritmética de las notas
promedio_notas = df['Nota'].mean()

# Mostramos por pantalla el promedio
print(f"--- Promedio de Notas ---")
print(f"{promedio_notas:.2f}")
```
