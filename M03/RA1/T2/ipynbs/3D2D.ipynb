```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(42)

n = 150

x0 = np.random.normal(0, 1, n)
y0 = np.random.normal(0, 1, n)
z0 = np.random.normal(-3, 0.8, n)

x1 = np.random.normal(0, 1, n)
y1 = np.random.normal(0, 1, n)
z1 = np.random.normal(3, 0.8, n)

X = np.concatenate([x0, x1])
Y = np.concatenate([y0, y1])
Z = np.concatenate([z0, z1])
labels = np.array([0]*n + [1]*n)
```

```python
fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X, Y, Z, c=labels, alpha=0.6)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Visualización en 3D (estructura real del dataset)")

plt.show()
```

```python
fig = plt.figure(figsize=(12,5))


plt.subplot(1,2,1)
plt.scatter(X, Y, c=labels, alpha=0.6)
plt.title("2D: X vs Y")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1,2,2)
plt.scatter(X, Z, c=labels, alpha=0.6)
plt.title("2D: X vs Z")
plt.xlabel("X")
plt.ylabel("Z")

plt.show()
```

```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.DataFrame({
    "X": X,
    "Y": Y,
    "Z": Z,
    "Clase": labels.astype(str)
})
```

```python
fig = px.scatter_3d(
    df,
    x="X",
    y="Y",
    z="Z",
    color="Clase",
    opacity=0.7,
    title="Scatter 3D interactivo (rotar con el ratón)"
)

fig.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame.copy()
df["species"] = df["target"].map(dict(enumerate(iris.target_names)))

x2d = "sepal length (cm)"
y2d = "sepal width (cm)"

plt.figure(figsize=(6,6))
for sp in df["species"].unique():
    sub = df[df["species"] == sp]
    plt.scatter(sub[x2d], sub[y2d], alpha=0.7, label=sp)

plt.xlabel(x2d)
plt.ylabel(y2d)
plt.title("Iris - Scatter 2D")
plt.legend()
plt.show()

x3d = "sepal length (cm)"
y3d = "sepal width (cm)"
z3d = "petal length (cm)"

fig = px.scatter_3d(
    df,
    x=x3d, y=y3d, z=z3d,
    color="species",
    opacity=0.75,
    title="Iris - Scatter 3D interactivo (rotar con el ratón)"
)
fig.show()
```
