```python
# Control de velocidad de un ventilador según la temperatura

def temp_fria(t):
    """Función de pertenencia 'fría'."""
    if t <= 15:
        return 1.0
    elif 15 < t < 20:
        return (20 - t) / 5.0
    else:
        return 0.0

def temp_templada(t):
    """Función de pertenencia 'templada' (triangular aprox. entre 18 y 27)."""
    if t <= 18 or t >= 27:
        return 0.0
    elif 18 < t < 22:
        return (t - 18) / 4.0
    elif 22 <= t < 27:
        return (27 - t) / 5.0
    else:
        return 0.0

def temp_caliente(t):
    """Función de pertenencia 'caliente'."""
    if t <= 24:
        return 0.0
    elif 24 < t < 30:
        return (t - 24) / 6.0
    else:
        return 1.0
```

```python
def sistema_difuso(temp):
    print(f"\n=== Sistema difuso para temperatura = {temp} ºC ===")

    mu_fria = temp_fria(temp)
    mu_templada = temp_templada(temp)
    mu_caliente = temp_caliente(temp)

    print("\n1) Fuzzificación (grados de pertenencia):")
    print(f"   μ_fria     = {mu_fria:.3f}")
    print(f"   μ_templada = {mu_templada:.3f}")
    print(f"   μ_caliente = {mu_caliente:.3f}")

    # Regla 1: SI temp ES fría → velocidad ES lenta (1)
    # Regla 2: SI temp ES templada → velocidad ES media (2)
    # Regla 3: SI temp ES caliente → velocidad ES rápida (3)

    fuerza_regla1 = mu_fria      # velocidad lenta
    fuerza_regla2 = mu_templada  # velocidad media
    fuerza_regla3 = mu_caliente  # velocidad rápida

    print("\n2) Activación de reglas:")
    print(f"   Regla 1 (fría → lenta):   fuerza = {fuerza_regla1:.3f}")
    print(f"   Regla 2 (templada → media): fuerza = {fuerza_regla2:.3f}")
    print(f"   Regla 3 (caliente → rápida): fuerza = {fuerza_regla3:.3f}")

    v_lenta = 1.0
    v_media = 2.0
    v_rapida = 3.0

    num = (fuerza_regla1 * v_lenta +
           fuerza_regla2 * v_media +
           fuerza_regla3 * v_rapida)
    den = (fuerza_regla1 + fuerza_regla2 + fuerza_regla3)

    if den == 0:
        velocidad_crisp = 0
    else:
        velocidad_crisp = num / den

    print("\n3) Defusificación (media ponderada):")
    print(f"   Numerador = {num:.3f}")
    print(f"   Denominador = {den:.3f}")
    print(f"   Velocidad resultante (crisp) = {velocidad_crisp:.3f}")

    if velocidad_crisp < 1.5:
        etiqueta = "lenta"
    elif velocidad_crisp < 2.5:
        etiqueta = "media"
    else:
        etiqueta = "rápida"

    print(f"\n==> Interpretación: velocidad {etiqueta.upper()} ≈ {velocidad_crisp:.2f}\n")

for t in [16, 20, 22, 26, 30]:
    sistema_difuso(t)
```

```python
# Mini clasificador de sentimiento con NLP muy básico

import re

PALABRAS_POSITIVAS = [
    "bien", "bueno", "genial", "fantastico", "increible",
    "contento", "feliz", "mejor", "encanta", "gusta"
]

PALABRAS_NEGATIVAS = [
    "mal", "malo", "horrible", "fatal", "triste",
    "odio", "terrible", "asco", "asqueroso", "frustrante"
]
```

```python

def limpiar_texto(texto):
    """Pasa a minúsculas y quita signos de puntuación."""
    texto = texto.lower()
    texto = re.sub(r"[^a-záéíóúñ0-9\s]", "", texto)
    return texto

def tokenizar(texto):
    """Divide el texto en palabras."""
    return texto.split()

def analizar_sentimiento(frase):
    print(f"\n=== Analizando frase: {frase!r} ===")

    limpia = limpiar_texto(frase)
    print("\n1) Texto normalizado:")
    print("   ", limpia)

    tokens = tokenizar(limpia)
    print("\n2) Tokens:")
    print("   ", tokens)
    pos = 0
    neg = 0
    for palabra in tokens:
        if palabra in PALABRAS_POSITIVAS:
            pos += 1
        if palabra in PALABRAS_NEGATIVAS:
            neg += 1

    print("\n3) Conteo de palabras:")
    print(f"   Palabras positivas encontradas: {pos}")
    print(f"   Palabras negativas encontradas: {neg}")

    score = pos - neg
    if score > 0:
        sentimiento = "POSITIVO"
    elif score < 0:
        sentimiento = "NEGATIVO"
    else:
        sentimiento = "NEUTRO"

    print("\n4) Resultado:")
    print(f"   Score = {score}")
    print(f"   Sentimiento estimado: {sentimiento}\n")
```

```python
frases = [
    "La clase de hoy ha sido genial, estoy muy contento.",
    "Este proyecto es horrible, estoy triste y frustrante.",
    "El ejercicio está bien, pero podría ser mejor.",
    "No sé qué pensar, es un día normal."
]

for f in frases:
    analizar_sentimiento(f)
```
